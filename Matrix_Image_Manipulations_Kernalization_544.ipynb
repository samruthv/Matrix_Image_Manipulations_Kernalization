```python
#***HW promlem 1***
import numpy as np
import math

def trace1D(list):
    
    #Making the list into a square matrix, if list the list can not be made into squre matirx ends code
    try:
        n = int(math.sqrt(len(list)))
        matrix = np.reshape(list, (n,n))
    except ValueError:
        print("not a square matix")
        return
        
    #Finding Trace of matrix
    trace=0
    for i in range(0,len(matrix)):
        
        diag=matrix[i][i]
        trace = trace + diag
    
    #Output   
    print("Trace of the suggested matrix is:{}".format(trace))
    return
    
  
    

trace1D(np.arange(0,100,1))
#print("Trace of the suggested matrix is: {}".format(trace1D(np.arange(0,100,1))))
```

### Part b

```python
#***HW promlem 2b***
dim = myRGB.shape
height, width, color = dim
#first filter
k=[[1,0,-1],[0,0,0],[-1,0,1]]
plt.imshow(k)
#looping through the hight and width and muliply filter to create new image
filteredPic = np.zeros(np.add(myRGB.shape, tuple([0,0,0])))
for i in range(height-2):
    for j in range(width-2):
        window = (myRGB[i:i+3, j:j+3])
        output_values = np.sum(k*window, axis=(0,1))
        filteredPic[i,j]=output_values

#display new image 

plt.imshow(filteredPic.astype('uint8'))
plt.show()
    
```

```python
#second filter
k=np.array([[0,-1,0],[-1,4,-1],[0,-1,0]])

#Same as above with new filter
filteredPic = np.zeros(np.add(myRGB.shape, tuple([0,0,0])))
for i in range(height-2):
    for j in range(width-2):
        window = (myRGB[i:i+3, j:j+3])
        output_values = np.sum(k*window, axis=(0,1))
        filteredPic[i,j]=output_values

#display new image 
plt.axis('off')
plt.imshow(filteredPic.astype('uint8'))
plt.show()
```

```python
def convolution2d(image, kernel, bias):
    m, n = kernel.shape
    if (m == n):
        y, x = image.shape
        y = y - m + 1
        x = x - m + 1
        new_image = np.zeros((y,x))
        for i in range(y):
            for j in range(x):
                new_image[i][j] = np.sum(image[i:i+m, j:j+m]*kernel) + bias
    return new_image
```

```python
#second filter
k=np.array([[1,0,-1],[0,0,0],[-1,0,1]])
plt.imshow(k)
plt.show()
#Same as above with new filter
filteredPic = np.zeros(np.add(myRGB.shape, tuple([0,0,0])))
for i in range(height-2):
    for j in range(width-2):
        window = (myRGB[i:i+3, j:j+3])
        output_values = np.sum(k*window, axis=(0,1))
        filteredPic[i,j]=output_values

#display new image 
plt.axis('off')
plt.imshow(filteredPic.astype('uint8'))
plt.show()
```

```python
#***HW promlem 3***

import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt

#load data
iris = datasets.load_iris().data

#normalize data 
means = np.mean(iris, axis=0)
std = np.std(iris, axis=0)
z_scores = (iris-means)/std

```

### Data Visualitaion of Iris Columns and Normalized Columns

```python
plt.plot(iris)
plt.show()
plt.plot(z_scores)
plt.show()
```
